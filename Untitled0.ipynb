```python
def Gnawed_Channel():
    Ligou_data = {
        '定位': '在小腿內側，當足內踝尖上5寸，脛骨內側面的中央。',
        '解剖': '在脛骨內側面下三分處；其內後側有大隱靜脈；布有隱神經的前支。',
        '主治': ' 月經不調，赤白帶下，陰挺，陰癢，疝氣，小便不利，睾丸腫痛，小腹痛，腰背拘急不可俯仰，脛部酸痛。',
        '配伍': ' 配百蟲窩、陰陵泉、三陰交治滴蟲性陰道炎；配中都、地機、中極、三陰交治月經不調、帶下症、睾丸炎；配大敦、氣沖治睾腫、卒疝、赤白帶下。',
        '刺灸法': '平刺0.5～0.8寸,可灸。'
    }
    return Ligou_data

def Pool_Spring():
    Ququan_data = {
        '定位': '在膝內側，屈膝，當膝關節內側端，股骨內側髁的後緣，半腱肌、半膜肌止端的前緣凹陷處。',
        '解剖': '在脛骨內髁後緣，半膜肌、半腱肌止點前上方；有大隱靜脈，膝最上動脈；布有隱神經、閉孔神經，深向膕窩可及脛神經。',
        '主治': '月經不調，痛經，白帶，陰挺，陰癢，產後腹痛，遺精，陽痿，疝氣，小便不利，頭痛，目眩，癲狂，膝臏腫痛，下肢痿痹。',
        '配伍': '配丘墟、陽陵泉治膽道疾患；配肝俞、腎俞、章門、商丘、太沖治肝炎；配復溜、腎俞、肝俞治肝腎陰虛之眩暈、翳障眼病；配支溝、陽陵泉治心腹疼痛、乳房脹痛、疝痛；配歸來、三陰交治肝鬱氣滯之痛經、月經不調。',
        '刺灸法': '直刺1～1.5寸,可灸。'
    }
    return Ququan_data

def Foot_Governor_of_Tears():
    Zuwuli_data = {
        '定位': '在大腿內側，當氣沖直下3寸，大腿根部，恥骨結節的下方，長收肌的外緣。',
        '解剖': '有內收長肌，內收短肌；有股內側動脈淺支；布有閉孔神經淺支和深支。',
        '主治': '少腹脹痛，小便不通，陰挺，睾丸腫痛，嗜臥，四肢倦怠，頸癧。',
        '配伍': '配三陽絡、天井、歷兌、三間治嗜臥欲動搖。',
        '刺灸法': '直刺0.5～0.8寸,可灸。'
    }
    return Zuwuli_data

def Cyclic_Gate():
    Qimen_data = {
        '定位': '在胸部，當乳頭直下，第6肋間隙，前正中線旁開4寸。',
        '解剖': '有腹直肌，肋間肌；有肋間動、靜脈；布有第六、七肋間神經。',
        '主治': '胸脅脹滿疼痛，嘔吐，呃逆，吞酸，腹脹，泄瀉，飢不欲食，胸中熱，咳喘，奔豚，瘧疾，傷寒熱入血室。',
        '配伍': '配大敦治疝氣；配肝俞、公孫、中脘、太沖、內關治肝膽疾患、膽囊炎、膽結石及肝氣鬱結之脅痛、食少、乳少、胃痛、嘔吐、呃逆、食不化、泄瀉等。',
        '刺灸法': '斜刺0.5～0.8寸；可灸。'
    }
    return Qimen_data

def Great_and_Thick():
    Dadun_data = {
        '定位': '在足大指末節外側，距趾甲角0.1寸。',
        '解剖': '有足趾背動、靜脈；布有腓神經的趾背神經。',
        '主治': '疝氣，縮陰，陰中痛，月經不調，血崩，尿血，癃閉，遺尿，淋疾，癲狂，癇證，少腹痛。',
        '配伍': '配內關、水溝治癲、狂、癇和中風昏仆；配膻中、天突、間使治梅核氣。',
        '刺灸法': '斜刺0.1～0.2寸，或用三陵針點刺出血；可灸。'
    }
    return Dadun_data

def Liver_Meridian_of_Foot_Jueyin():
    acupoints_info = {
        '蠡溝穴': Gnawed_Channel(),
        '曲泉穴': Pool_Spring(),
        '足五里穴': Foot_Governor_of_Tears(),
        '期門穴': Cyclic_Gate(),
        '大敦穴': Great_and_Thick()
        # ... 其他穴位
    }
    return acupoints_info

def get_acupoint_info(acupoint_name, info_category=None):
    acupoints_info = Lung_Meridian_of_Hand_Taiyin()

    acupoint_data = acupoints_info.get(acupoint_name, None)
    if acupoint_data is None:
        return "未找到指定的穴位信息。"

    if info_category:
        return acupoint_data.get(info_category, "未找到指定類別的信息。")

    return acupoint_data

def query_acupoint():
    while True:
        print("\n--- 查詢系統 ---")
        acupoint_name = input("請輸入您要查詢的穴位名稱（輸入'退出'結束查詢）:\n(蠡溝穴、曲泉穴、足五里穴、期門穴、大敦穴選擇其一owo):")
        if acupoint_name == "退出":
            print("\n--- 已退出查詢 ---")
            print("感謝使用本系統，祝你好運owo")
            break

        acupoint_info = get_acupoint_info(acupoint_name)
        if "未找到" in acupoint_info:
            print(f"\n{acupoint_info}\n")
            continue

        while True:
            print(f"\n找到穴位：{acupoint_name}")
            print("可查詢的信息類別有：定位、解剖、主治、配伍、刺灸法")
            info_category = input("請輸入您要查詢的信息類別（輸入'全部'顯示所有信息，或輸入'返回'返回上一層）:")

            if info_category == "返回":
                print("\n返回上一層...\n")
                break
            elif info_category == "全部":
                print(f"\n{acupoint_name} - 全部信息:")
                for category, detail in acupoint_info.items():
                    print(f"{category}: {detail}")
                print()  # 為了排版整潔owo
            elif info_category in ["定位", "解剖", "主治", "配伍", "刺灸法"]:
                category_info = get_acupoint_info(acupoint_name, info_category)
                print(f"\n{acupoint_name} - {info_category}:\n{category_info}\n")
            else:
                print("\n輸入的信息類別不正確，請重新輸入。\n")

            # 再次查詢或返回
            while True:
                next_action = input("輸入'再查詢'以查詢此穴位的其他信息，或輸入'返回'返回穴位選擇：")
                if next_action in ["再查詢", "返回"]:
                    break
                else:
                    print("請輸入有效的指令（'再查詢' 或 '返回'）。")
            if next_action == "返回":
                print("\n返回穴位選擇...")
                break
# 啟動查詢
query_acupoint()
```
